```python
import pandas as pd
import numpy as np
import pickle
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
df.info()
```

```python
df['diabetes'].value_counts()
```

```python
df.isna().sum()
```

```python
df['gender']=df['gender'].replace('female',1)
df['gender']=df['gender'].replace('male',2)
```

```python
df['diabetes']=df['diabetes'].replace('No diabetes',0)
df['diabetes']=df['diabetes'].replace('Diabetes',1)
```

```python
df.info()
```

```python
X=df[['cholesterol','glucose','hdl_chol','chol_hdl_ratio','age','gender','height','weight','bmi','systolic_bp','diastolic_bp','waist','hip','waist_hip_ratio']]
```

```python
X
```

```python
Y=df['diabetes']
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
#Training
clf = MultinomialNB()
clf.fit(X, Y)
```

```python
#Testing
clf.predict(np.array([232,184,114,2,91,1,61,127,24,170,82,35,38,0.92]).reshape(1,-1))
```

```python
#Testing
from sklearn.model_selection import train_test_split
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2) 
y_pred = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
pickle.dump(clf,open('diabetes_model.sav','wb'))
```
